```python
def find_min_max_in(df, col):
    """
    The function takes in a column and returns the top 5
    and bottom 5 in dataframe in that column.
    args:
        col: string - column name
    return:
        info_df: dataframe - final 5 dataframe
    """
    top = df[col].idxmax()
    top_df = pd.DataFrame(df.loc[top])
    bottom = df[col].idxmin()
    bottom_df = pd.DataFrame(df.loc[bottom])
    info_df = pd.concat([top_df, bottom_df], axis=1)
    return info_df

find_min_max_in(df, 'Total DL (Bytes)')

# This usually is helpful to understand various attributes of max and min values of a particular column. 
# This can also help to decide which column to keep for further analysis.
```

## Step 3) Data Cleaning/Manipulation


#### Fixing Columns
##### Dropping Columns with Single Value

```python

db_single=df.copy()

single_value_columns = pd.DataFrame(db_single.apply(lambda x: len(x.value_counts()), axis=0), columns=['SingleValueColumn'])

drop_single_value_columns = list(single_value_columns.loc[single_value_columns['SingleValueColumn']==1].index)

print('Columns which have just a single value => \n\n' + str(drop_single_value_columns))

df.drop(drop_single_value_columns, axis=1, inplace=True)

print('\n\nRemaining Columns => ' + str(len(df.columns)))
```

#### Dropping Columns with All Different Values


```python
db_diff=df.copy()

alldiff_value_columns = pd.DataFrame(db_diff.apply(lambda x: len(x.value_counts()), axis=0), columns=['AllDiffValueColumn'])

drop_alldiff_value_columns = list(alldiff_value_columns.loc[alldiff_value_columns['AllDiffValueColumn']==len(df.index)].index)

print('Columns which have all values different => \n\n' + str(drop_alldiff_value_columns))

df.drop(drop_alldiff_value_columns, axis=1, inplace=True)

print('\n\nRemaining Columns => ' + str(len(df.columns)))
```

#### Duplicate Data

```python
# Count of duplicates
df.duplicated().value_counts()
# Drop duplicates
df.drop_duplicates(keep='first', inplace=True)
df.drop_duplicates(subset=['MSISDN/Number'], keep='first', inplace=True)
```
